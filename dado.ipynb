```python
import pandas as pd
from IPython.display import display
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Supondo que 'dados' seja um DataFrame do pandas já carregado
# Exemplo de gráfico de dispersão entre 'população' e 'total_pontos'
plt.figure(figsize=(10, 6))

# Gráfico de dispersão
plt.scatter(dados['populacao'], dados['nome_uf'], alpha=0.6)

# Títulos e rótulos
plt.title('Relação entre População e Total de Pontos')
plt.xlabel('População')
plt.ylabel('Total de Pontos')

# Exibir grade
plt.grid(True)

# Mostrar gráfico
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Gráfico de barras
dados['nome_uf'].value_counts().plot.bar(color='skyblue')

# Títulos e rótulos
plt.title('Número de UFs')
plt.xlabel('Estados')
plt.ylabel('Contagem')

# Mostrar gráfico
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Histograma
dados['populacao'].plot.hist(bins=20, color='lightgreen', edgecolor='black')

# Títulos e rótulos
plt.title('Distribuição da População')
plt.xlabel('População')
plt.ylabel('Frequência')

# Mostrar gráfico
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Gráfico de linha
dados['populacao'].plot.line(color='orange')

# Títulos e rótulos
plt.title('Evolução da População')
plt.xlabel('Índice')
plt.ylabel('População')

# Mostrar gráfico
plt.grid(True)
plt.show()
```
